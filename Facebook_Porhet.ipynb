```python
import pandas as pd 
```

```python
#Read in the data 
df = pd.read_csv('/content/AAPL.csv')
df
```

```python
#Select the data and price 
df = df[['Date','Close']]
#Rename the features:
df = df.rename(columns={'Date':'ds','Close':'y'})
#Show the data 
df 
```

```python
#Get the last 20 rows of data store them into a new variable 
last = df[len(df)-20:]
last
```

```python
# Get all of the row expert the last 20 rows 
df = df [:-20]
df
```

```python
#Getting the Facebook Prophent library 
from fbprophet import Prophet
#Create the Prophent object(model)
fbp = Prophet(daily_seasonality = True)
#Fit or train the model 
fbp.fit(df)
future = fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
#Import the Facebooks prophet plotting libraries 
from fbprophet.plot import plot_plotly 
#plot the data 
plot_plotly(fbp,forecast)
```

```python
#Show the models prediction for 10-15-2021
forecast[forecast.ds =='2021-1-25']['yhat']
```
